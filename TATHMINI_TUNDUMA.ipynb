```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
# Loading the dataset
file_path_latest = './TATHMINI-YA-MUDA-UNAOTUMIKA-KWA-MAKASHA_cleaned.xlsx'
df_latest = pd.read_excel(file_path_latest)

# Renaming the columns
df_latest.rename(columns={
    "Lini Kasha liliondoka Bandari ya Dar es Salaam kuelekea nchi jirani?": "departure_date",
    "Lini kasha lilifika Mpakani kutokea Dar es Salaam?": "arrival_border_date",
    "Lini kasha liliondoka Mpakani kuelekea nchi jirani?": "departure_border_date",
    "Lini kasha lilifika kwa muagizaji wa shehena aliyeko nchi jirani?": "arrival_consignee_date",
    "Lini kazi ya kupakua shehena ilikamilika huko nchi jirani?": "unloading_completion_date",
    "Lini kasha lilifika mpakani likirejea Dar es Salaam?": "return_border_date",
    "Lini kasha linatarajiwa kufika Dar es Salaam?": "expected_arrival_date"
}, inplace=True)

# Convert date columns to datetime
date_columns_latest = [
    "departure_date", "arrival_border_date", "departure_border_date",
    "arrival_consignee_date", "unloading_completion_date",
    "return_border_date", "expected_arrival_date"
]

for col in date_columns_latest:
    df_latest[col] = pd.to_datetime(df_latest[col], errors='coerce')

# Calculate time taken for each leg of the journey
df_latest['Dar es Salaam to Border'] = (df_latest['arrival_border_date'] - df_latest['departure_date']).dt.days
df_latest['Border to Consignee'] = (df_latest['arrival_consignee_date'] - df_latest['departure_border_date']).dt.days
df_latest['Unloading of cargo'] = (df_latest['unloading_completion_date'] - df_latest['arrival_consignee_date']).dt.days
df_latest['Return to the Border'] = (df_latest['return_border_date'] - df_latest['unloading_completion_date']).dt.days
df_latest['Border to Dar es Salaam'] = (df_latest['expected_arrival_date'] - df_latest['return_border_date']).dt.days
df_latest['Total Turnaround Time'] = (df_latest['expected_arrival_date'] - df_latest['departure_date']).dt.days

# Display the calculated metrics
metrics_df_latest = df_latest[['Dar es Salaam to Border', 'Border to Consignee', 'Unloading of cargo', 'Return to the Border', 'Border to Dar es Salaam', 'Total Turnaround Time']]
metrics_summary_latest = metrics_df_latest.describe().round(0)

metrics_summary_latest
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Plotting the box plot with transposed axes
plt.figure(figsize=(10, 6))
metrics_df_latest.boxplot(vert=False)
plt.xlabel('Time (Days)')
plt.ylabel('Stages of the Journey')
plt.grid(False)  # Remove gridlines
plt.show()
```

```
# Remove outliers using the IQR method
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Remove outliers from the metrics dataframe
metrics_df_cleaned = remove_outliers(metrics_df_latest, metrics_df_latest.columns)

# Save the cleaned dataset to a CSV file
cleaned_filename = 'TATHMINI_DSM_DRC_ZAMBIA.csv'
metrics_df_cleaned.to_csv(cleaned_filename, index=False)

# Display the summary of the cleaned metrics
metrics_summary_cleaned = metrics_df_cleaned.describe().round(0)
metrics_summary_cleaned
```

```
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Define stopwords (if Swahili is not available in NLTK, we may need a custom list)
swahili_stopwords = [
    "na", "ya", "kwa", "wa", "wao", "ni", "la", "hiyo", "katika", "kuwa",
    "kama", "mimi", "sisi", "yeye", "wote", "kila", "moja", "mbili",
    "mbele", "nyuma", "juu", "chini", "ndani", "nje", "kati", "zaidi",
    "chini", "wote", "pia", "lakini", "ama", "hata", "kwa", "huo", "hivyo"
]

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to preprocess text
def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    # Convert to lower case and remove stopwords and punctuation
    tokens = [word.lower() for word in tokens if word.isalnum() and word.lower() not in swahili_stopwords]
    # Lemmatize the tokens
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return tokens

# Extract the relevant column and drop null or empty entries
text_data = df_latest['Maelezo mengine ya ziada kuhusu muda unaotumika hadi kurejea nchini makasha yaliyoelekea nchi jirani na sababu zake'].dropna().reset_index(drop=True)

# Apply preprocessing to the text data
preprocessed_text = text_data.apply(preprocess_text)

# Display the first few preprocessed entries
preprocessed_text.head()
```

```
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Flatten the list of tokens
all_tokens = [token for sublist in preprocessed_text for token in sublist]

# Count the frequency of each word
word_freq = Counter(all_tokens)

# Display the most common words
print(word_freq.most_common(20))

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
