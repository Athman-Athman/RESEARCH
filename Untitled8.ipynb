```python
import matplotlib.pyplot as plt
import seaborn as sns
country_quantity = data.groupby('ORIGIN COUNTRY')['QUANTITY'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='ORIGIN COUNTRY', y='QUANTITY', data=country_quantity)
plt.xticks(rotation=90)
plt.title('QUANTITY BY COUNTRY')
plt.xlabel('ORIGIN COUNTRY')
plt.ylabel('QUANTITY')
plt.tight_layout()
plt.show()
```
